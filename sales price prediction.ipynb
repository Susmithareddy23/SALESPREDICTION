```python
columns = ['Sales', 'Newspaper', 'Radio','TV']
for column in columns:
    sns.violinplot(x='TV', y=column, data=sales)
    plt.xlabel('TV')
    plt.ylabel(column)
    plt.title(f'{column} vs TV')
    plt.show()
```

```python
columns = ['Sales', 'Newspaper', 'Radio','TV']
for column in columns:
    sns.boxplot(x='TV', y=column, data=sales)
    plt.xlabel('TV')
    plt.ylabel(column)
    plt.title(f'{column} vs TV')
    plt.show()
```

```python
columns = ['Sales', 'Newspaper', 'Radio','TV']
for column in columns:
    sns.lineplot(x='TV', y=column, data=sales)
    plt.xlabel('TV')
    plt.ylabel(column)
    plt.title(f'{column} vs TV')
    plt.show()
```

```python
columns = ['Sales', 'Newspaper', 'Radio','TV']
for column in columns:
    sns.swarmplot(x='TV', y=column, data=sales)
    plt.xlabel('TV')
    plt.ylabel(column)
    plt.title(f'{column} vs TV')
    plt.show()
```
